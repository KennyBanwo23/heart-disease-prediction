### Train Test Split

```python
# Classification Report of Logistic Regression
target_name = 'target'

# Separate object for target feature
y = df[target_name]

# Separate object for input features
X = df.drop(target_name, axis = 1)
```

```python
# Apply Standard Scaler
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X)
SSX = scaler.transform(X)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(SSX, y, test_size = 0.15, random_state = 7) 
```

```python
# Communication 
def make_predictionRandom(age, sex, cp, trestbps, chol, fbs, restecg, thalach, exang, oldpeak, slope, ca, thal):
    data = {
         'age' : age,
        'sex' : sex,
        'cp': cp, 
        'trestbps' : trestbps,
        'chol' : chol, 
         'fbs' : fbs, 
         'restecg' : restecg, 
         'thalach' : thalach, 
        'exang' : exang,
        'oldpeak' : oldpeak,
        'slope' : slope,
        'ca' : ca,
        'thal' : thal
    }   
    df = pd.DataFrame(data, index = [0])
    
    rf_pred = rf.predict(df).round(2)[0]
    if rf_pred == 1:
        return f"Predicted Diagnosis: {rf_pred}. This patient has heart disease."
    else:
        return f"Predicted Diagnosis: {rf_pred}. This patient does not have heart disease."
        
```
